## Separacion de los datos

Para tener distribuciones similares en los datos de entrenamiento y los de prueba, implementaremos stratified suffle split:

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

Lo que estamos haciendo aqui es asegurar que se mantengan distribuidas de manera similar las variables de importancia para el problema

```python
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2)
for train_indices, test_indices in split.split(titanic_data, titanic_data[["Survived", "Pclass", "Sex"]]):
    strat_train_set = titanic_data.loc[train_indices]
    strat_test_set = titanic_data.loc[test_indices]
```

Para observar las distribuciones podemos representarlas en los siguientes plots:

Al analizar, podemos observar que existen una gran proporcion de valores faltantes en la variable sexo en set de entrenamiento. Para resolver esto podemos inputar los valores faltantes. 

Existen varias formas: 
- Inputation 
- Dropping features 
- Feature encoding
- Estimators

To fill the NaN values we can use the mean, use random values

```python
strat_train_set.info()
```

Calculate the estimators:

```python
# Transforme mixin nos permitira automaticamente mezclar fit and transform 
# dentro de la funcion fit_transform sin tener de definirlo explicitamente 
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.impute import SimpleImputer # how to fill the na values

class AgeImputer(BaseEstimator, TransformerMixin):
    # it is used by BaseEstimator
    def fit(self,X, y=None):
        return self
    
    def transform(self, X):
        imputer = SimpleImputer(strategy="mean")
        X['Age'] = imputer.fit_transform(X[['Age']])
        return X
```

Despues queremos encode determidas variables. One hot encoding is called in machine learning. 
We have a feature with three possible values which is one column, instead of having that we're transforming that into three binary features

```python
from sklearn.preprocessing import OneHotEncoder

class FeatureEncoder(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        encoder = OneHotEncoder()
        matrix = encoder.fit_transform(X[["Embarked"]]).toarray()
        
        column_names = ["C", "S", "Q", "N"]
        
        for i in range(len(matrix.T)):
            X[column_names[i]] = matrix.T[i]
            
        matrix = encoder.fit_transform(X[['Sex']]).toarray()
        
        column_names = ["Female", "Male"]
        
        for i in range(len(matrix.T)):
            X[column_names[i]] = matrix.T[i]
            
        return X
```

```python
class FeatureDropper(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        return X.drop(["Embarked", "Name", "Ticket", "Cabin", "Sex", "N"], axis=1, errors='ignore')
```

```python
from sklearn.pipeline import Pipeline

pipeline = Pipeline([('ageimputer', AgeImputer()),
                     ('featureencoder', FeatureEncoder()),
                     ('featuredropper', FeatureDropper())])

#pipeline = Pipeline([('ageimputer', AgeImputer())])
```

```python
strat_train_set = pipeline.fit_transform(strat_train_set)
```

```python
strat_train_set.info()
```

We have to scale all of this and then we're going to start with the actual training

```python
from sklearn.preprocessing import StandardScaler

X = strat_train_set.drop(['Survived'], axis=1)
Y = strat_train_set['Survived']

scaler = StandardScaler()
X_data = scaler.fit_transform(X)
# because we won't scale the survival but the numerical feature should 
# be scaled because that usually makes it easier for the models
Y_data = Y.to_numpy() # we transform it because the X_data is a numpu array
```

```python
X_data
```

## Train the model 

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV # is cv stands for cross validation
# the idea is you have 100 percent of your data and you split it into 10 folds for example 
# this is what it's called folds. What you do is you take nine of those folds train them all on 
# those nine and you test it you evaluate it on the one fold that;s remaining
# and you do that with all possible combinations so you take all different variation of that 
# so that there's always each of these folds was used once for validation. 
# This is what you do to validate your model so you don't do that with a test data set. 
# you do that inside of the training data set you split it into validation data sets and the GridSearch uses cross validation
# so what i just explainded to find the optimal hyper parameters 

clf = RandomForestClassifier()

# the following hyperparameters will be used to build all the possible combinations and test them
param_grid = [
    {'n_estimators': [10, 100, 200, 500], "max_depth": [None, 5, 10], "min_samples_split": [2, 3, 4]}
]

grid_search = GridSearchCV(clf, param_grid, cv=3, scoring="accuracy", return_train_score=True)
grid_search.fit(X_data, Y_data)
```

```python
final_clf = grid_search.best_estimator_
```

```python
final_clf
```

### Probar el modelo en el set de test 

```python
strat_test_set = pipeline.fit_transform(strat_test_set)
```

```python
strat_test_set
```

```python
X_test = strat_test_set.drop(["Survived"], axis=1)
y_test = strat_test_set['Survived']

scaler = StandardScaler()
X_data_test = scaler.fit_transform(X_test)
y_data_test = y_test.to_numpy()
```
